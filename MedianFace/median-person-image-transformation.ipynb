```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# Load the os library
import os

# Load the request module
import urllib.request

# Import SSL which we need to setup for talking to the HTTPS server
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

# Create a directory
try: 
    os.mkdir('publicFace')

    # Now perform the following 100 times:
    for img_i in range(1, 101):

        # create a string using the current loop counter
        f = '000%03d.jpg' % img_i

        # and get the url with that string appended the end
        url = 'http://mmlab.ie.cuhk.edu.hk/projects/TCDCN.html' + f

        # We'll print this out to the console so we can see how far we've gone
        print(url, end='\r')

        # And now download the url to a location inside our new directory
        urllib.request.urlretrieve(url, os.path.join('publicFace', f))
except:
    #os.rm('img_align_celeba')
    print("You may need to delete the existing 'publicFace' folder in your directory")
```

```python
files = os.listdir('publicFace')# img.<tab>
import matplotlib.pyplot as plt
import numpy as np

print(os.path.join('publicFace', files[0]))
plt.imread(os.path.join('publicFace', files[0]))

files = []
for file_i in os.listdir('publicFace'):
    if '.jpg' in file_i:
        files.append(os.path.join('publicFace', file_i))

# There should be 102 files, with the last one being number 101

img = plt.imread(files[101])

print(img)
```

```python
imgs = [plt.imread(files[file_i])
        for file_i in range(102)]
```

```python
data = np.array(imgs) # make 'data' = our numpy array
data.shape
print(data.shape)
print("The shape of our new 'data' object is a 'batch' of 102 images, with a height of 150, width of 150, and 3 colour channels")
print("If your images aren't all the same size to begin with, then this won't work!")
```

```python
median_img = np.median(data, axis=0) # This is the median of the 'batch' channel
plt.imshow(median_img.astype(np.uint8))
print("look at this median person")
```
